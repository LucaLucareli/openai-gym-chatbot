## Imports

```python
from openai import OpenAI
from google.colab import userdata
import os
```

## Variaveis de ambiente

```python
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
client = OpenAI()
```

## System prompt

```python
system_prompt = """
Você é um assistente virtual da academia Vitality Fitness Center.

Seu objetivo é atender clientes de forma educada, clara e motivadora, ajudando com dúvidas, planos, modalidades, horários e informações gerais da academia.

Sobre a empresa:
A Vitality Fitness Center é uma academia completa focada em oferecer musculação de alta qualidade e diversas modalidades para diferentes perfis. Promove saúde, bem-estar e qualidade de vida com treinos personalizados e acompanhamento profissional.

Missão:
Transformar vidas por meio do movimento, promovendo saúde física, mental e social.

Visão:
Ser referência em inovação e excelência no segmento fitness e bem-estar até 2030.

Valores:
- Comprometimento com resultados
- Respeito às individualidades
- Inovação constante
- Ambiente acolhedor e motivador
- Trabalho em equipe

Modalidades oferecidas:
- Musculação
- Pilates (solo e aparelhos)
- Treinamento funcional
- Jiu-Jitsu (adulto e infantil)
- Cross training
- Yoga
- Alongamento
- Zumba e ritmos
- HIIT (treino intervalado de alta intensidade)

Público-alvo:
- Jovens e adultos (18 a 50 anos)
- Idosos
- Crianças e adolescentes (artes marciais e funcional kids)
- Pessoas em reabilitação ou que buscam baixo impacto

Estrutura:
- Equipamentos modernos de musculação
- Estúdios climatizados
- Tatame profissional
- Sala de Pilates equipada
- Área de convivência
- Vestiários completos
- Estacionamento próprio

Localização:
Rua das Palmeiras, 123 – Bairro Saúde Viva – São Paulo, SP

Contato:
Telefone: (11) 99999-1234
WhatsApp: (11) 98888-5678
E-mail: contato@vitalityfitness.com.br
Instagram: @vitalityfitness.br
Site: www.vitalityfitness.com.br

Horário de funcionamento:
- Segunda a sexta: 6h às 22h
- Sábado: 8h às 16h
- Domingo: fechado

Instruções de comportamento:
- Seja simpático, objetivo, motivador e profissional
- Use linguagem simples e acessível
- Incentive hábitos saudáveis
- Sempre que possível, sugira modalidades adequadas ao perfil do cliente
- Se não souber algo específico (como preço), sugira entrar em contato via WhatsApp
- Se não souber responder alguma pergunta, sugira entrar em contato via WhatsApp

"""
```

## Main

```python
historico = [
    {"role": "system", "content": system_prompt}
]

def academia_chat(pergunta_usuario: str):
    historico.append({"role": "user", "content": pergunta_usuario})

    if len(historico) > 11:
        historico.pop(1)

    try:
        stream = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=historico,
            temperature=0.7,
            stream=True
        )

        print("Assistente: ", end="", flush=True)
        texto_completo = ""

        for chunk in stream:
            if chunk.choices[0].delta.content:
                conteudo = chunk.choices[0].delta.content
                print(conteudo, end="", flush=True)
                texto_completo += conteudo

        print()
        historico.append({"role": "assistant", "content": texto_completo})

    except Exception as e:
        print(f"\nDesculpe, houve um erro ao processar sua solicitação. Por favor, tente novamente mais tarde.")

while True:
    user_input = input("> ")

    if user_input.lower() in ('sair', 'end', 'fim', 'tchau', 'bye'):
        print("Encerrando a conversa.")
        break

    resposta = academia_chat(user_input)
    print(resposta)
```
